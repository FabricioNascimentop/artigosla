```python
import pandas as pd

df = pd.read_csv('macro_barcarena_tratado.csv')
df
```

```python
import plotly.express as px

fig = px.bar(df, x='ano',y='PIB')
fig.show()
```

```python
import plotly.graph_objects as go
from utils import formatar_moeda_br

df['Impostos_text'] = df['Impostos'] * 1000
df['Impostos_text'] = df['Impostos_text'].round(2)
df["Impostos_text"] = df["Impostos_text"].apply(
    lambda x: f"R$ {x:,.0f}".replace(",", "X").replace(".", ",").replace("X", ".")
)

df['vab_bruto_text'] = df['vab_bruto'] * 1000
df['vab_bruto_text'] = df['vab_bruto_text'].round(2)
df["vab_bruto_text"] = df["vab_bruto_text"].apply(
    lambda x: f"R$ {x:,.0f}".replace(",", "X").replace(".", ",").replace("X", ".")
)


df['Impostos_p'] = (df['Impostos'] / df['PIB'] * 100).round(2).astype(str) + '%'



fig = go.Figure(data=[
    go.Bar(x=df['ano'],y=df['Impostos'],name='Impostos',text=df['Impostos_text']),
    go.Bar(x=df['ano'], y=df['vab_bruto'],name='Valor adicionado bruto', text=df['vab_bruto_text'],textposition="outside",textfont=dict(color="black",size=25)),
    go.Scatter(x=df['ano'], y=df['Impostos'], mode='markers+text',text=df['Impostos_p'], textposition="top center",name="proporção imposto/pib",textfont=dict(color="white",size=10))
])

fig.update_layout(barmode='stack')
fig.show()
```
